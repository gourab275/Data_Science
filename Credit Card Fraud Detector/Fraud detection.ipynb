```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import xgboost as xgb
import lightgbm as lgb
# !pip install lightgbm
```

```python
legit = credit[credit.Class==0]
fraud = credit[credit.Class==1]
```

```python
# '''as the fradulant data is a 
# less than the legit data (imbalanced data) we have to under sample the data'''
```

```python
legit_sample = legit.sample(n=492)
```

```python
cred = pd.concat([legit_sample,fraud],axis=0)
```

```python
print(cred.shape)
cred.head()
```

```python
cred.Class.value_counts()
```

```python
cred.groupby('Class').mean()
```

```python
x = cred.drop(columns=['Class'],axis=1)
y = cred.Class
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,stratify=y,random_state = 4)
```

# Model building 

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_predict = model.predict(x_train)
```

```python
acc_train = accuracy_score(x_train_predict,y_train)
print(acc_train)
```

```python
x_test_predict = model.predict(x_test)
acc_test = accuracy_score(x_test_predict,y_test)
print(acc_test)
```

```python
# XGBoost Model

xgb_clf = xgb.XGBClassifier(max_depth = 3, n_estimators = 100, 
                            learning_rate = 0.1, n_jobs = -1,
                           eval_metric='logloss',objective= 'binary:logistic')
```

```python
xgb_clf1 = xgb_clf.fit(x_train, y_train)
```

```python
xgb_pred = xgb_clf1.predict(x_test)
xgb_pred1 = xgb_clf1.predict(x_train)
```

```python
print(accuracy_score(y_train, xgb_pred1))
print(accuracy_score(y_test, xgb_pred))
```
